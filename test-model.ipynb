```python
# BIBLIOTHEQUES
from ultralytics import YOLO
import cv2
from PIL import Image
import os
```

```python
# CONFIGURATION
MODEL_PATH = "runs/detect/train2/weights/best.pt"  
IMAGE_PATH = "img-test/img2.jpg"  
CONFIDENCE = 0.30  
SAVE_PATH = "resultat.jpg"  
```

```python
# CHARGE LE MOD√àLE
if os.path.exists(MODEL_PATH):
    model = YOLO(MODEL_PATH)
    print("‚úÖ Mod√®le custom charg√© (mAP50: 69.4%)")
else:
    print("‚ùå Mod√®le non trouv√©, utilise YOLOv8n pr√©-entra√Æn√©")
    model = YOLO("yolov8n.pt")
```

```python
# V√âRIFIE L'IMAGE
if not os.path.exists(IMAGE_PATH):
    print(f"‚ùå Image '{IMAGE_PATH}' non trouv√©e!")
    print("üì§ Mets une image dans le dossier et relance")
    exit()
```

```python
# D√âTECTION
print("üîç D√©tection en cours...")
results = model(IMAGE_PATH, conf=CONFIDENCE, verbose=True)
```

```python
# AFFICHE R√âSULTATS
result_img = results[0].plot()  
result_pil = Image.fromarray(result_img)
```

```python
# Affiche original vs d√©tect√©
img_original = Image.open(IMAGE_PATH)
img_original.show(title="Image Originale")

result_pil.show(title="Objets D√©tect√©s (mAP50: 69.4%)")
```

```python
# SAUVEGARDE
result_pil.save(SAVE_PATH)
print(f"üíæ R√©sultat sauv√©: {SAVE_PATH}")
```

```python
# STATS D√âTAILL√âES
boxes = results[0].boxes
if boxes is not None:
    print("\nüéØ OBJETS D√âTECT√âS:")
    classes = model.names
    for box in boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        class_name = classes[cls_id]
        print(f"  ‚Üí {class_name}: {conf:.1%}")
else:
    print("‚ÑπÔ∏è Aucun objet d√©tect√© (baisse CONFIDENCE?)")
```
